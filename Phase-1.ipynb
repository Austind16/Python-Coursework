## Numpy array and basics
Learn about numpy

```python
import numpy as np
```

## Creating array from list

```python
arr_1d = np.array([1, 2, 3])
print("1D array is: ", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array is: ", arr_2d)
```

### List vs numpy array

```python
py_list = [1, 2, 3]
print("The multiplication the list is ", py_list * 2)

np_array = np.array([1, 2, 3]) # element wise multiplication
print("The array multiplication is ", np_array * 2)

import time
start = time.time()
py_list = [i *2 for i in range(100000)]
print("\n Time for py operation is: ",  time.time() - start)

import time
start = time.time()
numpy_list = np.arange(100000) * 2
print("\n Time for numpy operation is: ",  time.time() - start)
```

### Creating Array from scratch

```python
np_zeros = np.zeros((3, 4))
print("Zero Arrays: \n", np_zeros)

np_ones = np.ones((4, 3))
print("One Arrays: \n", np_ones)

np_constants = np.full((3, 3), 6)
print("Constant Arrays: \n", np_constants)

np_random = np.random.random((2, 2))
print("Random Arrays: \n", np_random)

np_sequence = np.arange(23, 54, 3)
print("Sequence Arrays: \n", np_sequence)
```

### Vector, Matrices and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3], 
                  [4, 5, 6]])
print("Matrix: ", matrix)

tensor = np.array([[[1, 2], [3,4]],
                  [[5, 6], [7, 8]]])
print("Tensor:", tensor)
```

### Array Properties

```python
np_array = np.array([[1, 2, 3], 
                     [4, 5, 6]])
print("Shape: ", np_array.shape)
print("Dimension: ", np_array.ndim)
print("Size: ", np_array.size)
print("Datatype: ", np_array.dtype)
```

### Array Reshaping

```python
original = np.arange(12)
print("Original array: ", original)

reshaped = original.reshape((3, 4))
print("\n Reshaped array: \n", reshaped)

flattened = reshaped.flatten()
print("\n Flattened array: \n", flattened)

# Ravel (returns new, instead of copy)
raveled = flattened.ravel()
print("\n Raveled array: \n", raveled)

transpose = reshaped.T
print("\n Transpose of array: \n", transpose)
```
