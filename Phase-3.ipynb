# Advanced Operations with bussiness examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # McDonalds
    [2, 120000, 140000, 160000, 190000], # KFC
    [3, 200000, 230000, 260000, 300000], # Faasos
    [4, 180000, 210000, 240000, 270000], # Burger King
    [5, 160000, 185000, 205000, 230000], # Dominoz pizza
])

print(" ==== Zomato Sales Analysis ==== ")
print("\n Sales Data Shape: ", np.shape(sales_data))
print("\n Sales Data of first three restaurants: \n", sales_data[0:3])
```

```python
# Total sales per year
print("Sales:\n", np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("Yearly Total: \n", yearly_total)

# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Yearly minimum sales:\n", min_sales)

# Maximum sales yearly
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Yearly maximum sales:\n", max_sales)

# Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Restaurant average sales:\n", avg_sales)

# Cumulative sales
cum_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Restaurant cumalative sales:\n", cum_sales)

plt.figure(figsize= (10, 6))
plt.plot(np.mean(cum_sales, axis=0))
plt.title("Average Cumulative sales across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)
print("Vector Multiplication: ", vector1 * vector2)
print("Vector Dot Product: ", np.dot(vector1, vector2))
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1 * np.linalg.norm(vector2))))
print("Angle is:", angle)
```
